```python
df1=df.dropna
df1
```

```python
#Duplicates 
df.drop_duplicates
```

```python
#renamiming the column
df.rename(columns={"search interest": "search_interest"}, inplace=True)
df1

```

```python
df.rename(columns={"increase percent": "increase_percent"}, inplace=True)
df1
```

```python
df
```

```python
#step 1:convert to numeric
df['increase_percent'] = df['increase_percent'].str.replace('%','',regex=False)
```

```python
#step 2:convert to numeric
df['increase_percent'] = pd.to_numeric(df['increase_percent'],errors='coerce')
df.dtypes
```

```python
#Remove Outliers (IQR Method for Numeric Columns)
Q1 =df['search_interest'].quantile(0.25)
Q3 =df['search_interest'].quantile(0.75)
IQR= Q3-Q1

print("Q1:",Q1)
print("Q3:",Q3)
print("IQR:",IQR)
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df1
    
```

```python
df['increase_percent'] = pd.to_numeric(df['increase_percent'], errors='coerce')
df
```

```python
print(df.info())
print(df.describe())
print("Final Shape:", df.shape)
```

```python
#detect outliers
outliers =df[(df['search_interest']< lower)|
             (df['search_interest']>upper)]

print("Outliers Detected:")
display(outliers)
```

```python
df['growth_category'] = np.where(
    df['increase_percent'] >100 , 'High Growth',
     np.where(df['increase_percent'] >= 50, 'Medium Growth','Low Growth')
 )
df
```

```python
correlation = df[['search_interest', 'increase_percent']].corr()
print(correlation)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['increase_percent'], df['search_interest'])
plt.xlabel("Increase percent")
plt.xlabel("Search intreset")
plt.title("Search interest vs Increase percent")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['search_interest'], bins=10)
plt.xlabel("search_Interest")
plt.ylabel("Frequency")
plt.title("Distribution of Search Interest")
plt.show()
```
